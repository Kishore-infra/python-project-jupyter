```python
import datetime
from collections import Counter

class Post:
    def __init__(self,post_id,title,contant,created_on):
        self.post_id = post_id
        self.title = title
        self.contant = contant
        self.created_on = created_on

    def __str__(self):
         return (
           f"Post_Id    : {self.post_id}\n"
           f"Title      : {self.title}\n"
           f"Contant    : {self.contant}\n"
           f"Created_on : {self.created_on}\n"
            "--------------------------------"
       )
        
      
   
class NotFoundError(Exception):
    pass
 
class Blog(Post):
    def __init__(self):
        self.__posts = []
        self.__counter = 0

    def create_post(self):
        self.__counter += 1
        post_id = self.__counter
        title = input("Enter Your Title:\n")
        contant = input("Enter Your Thoughts:\n")
        created_on = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        new_post = Post(post_id,title,contant,created_on)
        self.__posts.append(new_post)
        print("Post Created successfully!!")

    def view_post(self):
        if not self.__posts:
            print("No post available")
            return
        for post in self.__posts:
            print(f"""
Post_ID    : {post.post_id}
Title      : {post.title}
Contant    : {post.contant}
Created_on : {post.created_on}
----------------------------
""")

    def edit_post(self):
        post_id = int(input("Enter Your ID: \n"))
        post = next((p for p in self.__posts if p.post_id == post_id ),None)
        
        if post is None:
             raise NotFoundError("Invalied post ID")

        if(post):
            title = input("Enter The Title:\n")
            contant = input("Enter Your Thoughts:\n")
            post.title = title
            post.contant = contant
            print("Post Updated Successfully!!")

    def delete_post(self):
        post_id = int(input("Enter Your ID:\n"))
        post = next((p for p in self.__posts if p.post_id == post_id ),None)
        
        if post is None:
            raise NotFoundError("Invalied post ID")
       
        conform  = input("Are you sure want do delete (yes/no):").lower()
        if conform == "yes":
            self.__posts.remove(post)
            print("Post Deleted Successfully")
        else:
            print("Deletion Cancelled!")

  

            

    def export_file(self):
        if not self.__posts:
            print("Post Not Avaliable")
            return

        with open ("posts.txt","w") as f:
            for post in self.__posts:
                f.write(f"""
                    Post_ID    : {post.post_id}
                    Title      : {post.title}
                    contant    : {post.contant} 
                    Created_on : {post.created_on}     
                    -------------------------------
                    """)
            f.close()
            print("File Export Successfully!!!")


    def start(self):

        while True:
            try:
                print("Select Your options")
                print("1.create Post")
                print("2.Edit Post")
                print("3.View Post")
                print("4.Delete Post")
                print("5.Export Post Into File.txt")
                print("6.Application stopped")

                selected_option = int(input())

                if selected_option == 1:
                    self.create_post()
                elif selected_option == 2:
                    self.edit_post()
                elif selected_option == 3:
                    self.view_post()
                elif selected_option == 4:
                    self.delete_post()
                elif selected_option == 5:
                    self.export_file()
                elif selected_option == 6:
                    print("Application Stopped")
                    break
                else:
                    print("Invalied Choice!!")
                    continue

            except ValueError as e:
                print("Restarting Application")
                continue
            except NotFoundError as e:
                print(e)
                print("Restarting Application")
                continue
                


                    
                
                
        
```

```python
blog = Blog()
blog.start()
```
